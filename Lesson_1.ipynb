1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.
2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Когда лучше максимизировать Precision? Максимизировать Precision необходимо тогда, когда в результирующей модели важна точность, обусловленная бизнес задачей. Например если модель предсказывает диагноз, где стоимость ошибки жизнь пациента, или если модель предсказывает место где разместить магазин, в этом случае цена ошибки издержки(затраты на создание торговой точки, аренда, ремонт и т.д)

Когда лучше максимизировать Recall? Если модель предсказывает выйдет ли из стоя оборудование, то метрика  Recall будет кстати. Если модель предскажет, что оборудование на выйдет из стоя, а оно вышло из стоя, то цена такой ошибки высока(замена, ремонт, простой оборудования + простой части технологического процесса).  Метрику Recall необходимо максимизировать.


Мы используем F-меру потому, что эта мера показывает "перекос", если одна из метрик входящая в уравнение будет очень мала, то F-мера это покажет, тоже будет представлена малым значением. Если взять среднее от Precision и Recall, то "перекоса" мы не увидим.
